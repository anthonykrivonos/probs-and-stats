## Cumulative Distributions

The cumulative distribution function (CDF) of the r.v. $X$ is defined by

$ F(X) = P(X ≤ x) $ for $ x \in \mathbb{R} $

"Cumulative" implies that as $x \rightarrow \infty $, $F(x)$ grows.

### Notation

The notation $X ∼ F$ is used to denote that $F$ is the distribution function
for $X$. **This is very important**, as we will soon explore many popular
probability distributions.

This is very important.

### Example

Suppose we flip a coin. Our r.v. X can be any of the following:
$x \in \{ 1, 2, 3, 4, 5, 6 \}$. This will be our *support* or, simply,
the range of possible values.

There's only a $\frac{1}{6}$ chance of landing $x = 1$.

But, there's a $\frac{2}{6}=\frac{1}{3}$ chance of landing $x ≤ 2$!

And then there's a $\frac{3}{6}=\frac{1}{2}$ chance of landing $x ≤ 3$!

This is how cumulative distributions work, and they can be modeled as follows.

```python
%matplotlib inline

import matplotlib.pyplot as plt

x = [ 1, 2, 3, 4, 5, 6 ]
y = list(map(lambda val: val/6, x))

plt.step(x, y)
plt.show()
```

## Discrete Random Variables and Probability Mass Functions

These r.v.'s have supports that are sequential. In other words,

For a discrete r.v. $X$, we define the probability mass function (PMF)
of $X$ by

$ \boxed{p(a)$ = $P(X=a)}$

The probability mass function p(a) is positive for at most a
countable number of values of $a$. That is, for some $a_i$, $p(a_i)=0$.

Since the sum of the probabilities of all $x_i$ is 1 (by the second axiom
of probability), we have that:

$\sum_{i=1}^{\infty} p(x_i) = 1$

Most importantly, the PMF of $X$ is simply the derivative of its CDF:

$ \boxed{f(a) = \frac{d}{da}F(a)}$

### Example

Take our six-sided fair die example from above.

The probability of landing a 1 is $\frac{1}{6}$. The probability of landing a
2 is $\frac{1}{6}$. The probability of landing a 3 is $\frac{1}{6}$, and so on.

So, since probability of landing any number is $\frac{1}{6}$, our PMF will intuitively look like a
horizontal line with $f(x) = \frac{1}{6}$ on $ x \in [1, 6]$.

## Jointly Distributed Random Variables

To discern the relationship between two random variables $X$ and $Y$,
we denote their joint distribution function:

$ \boxed{F(x, y) = P(X ≤ x, Y ≤ y)} $

If $X$ and $Y$ are independent, then

$ F(x, y) = P(X ≤ x) × P(Y ≤ y)$

You can get the *marginal distribution* of $X$ and $Y$ separately by
"covering the entire support" of the opposite variable. In other words,
to find the marginal distribution of $X$, $F_X(x)$, we simply find
$F(x, \infty) = P(X ≤ x, Y ≤ \infty)$.

This can be done via integration:

$ F_X(x) = \int_{-\infty}^{\infty}{F(x, y)dy} $

$ F_Y(x) = \int_{-\infty}^{\infty}{F(x, y)dx} $

Note: The above definition for joint distributions can be extended to any
number of r.v.s $X_1, X_2, ..., X_n$!

### Example

The 3D plot below is an example of a joint CDF for r.v.s $X$ and $Y$.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def arb_cdf_func(x, y):
    return x + y

def pdf(start, end, func):
    x = y = np.arange(start, end + 1, 1)
    X, Y = np.meshgrid(x, y)
    f = func(np.ravel(X), np.ravel(Y))
    F = np.array(f).reshape(X.shape)
    return X, Y, F

X, Y, F = pdf(1, 6, arb_cdf_func)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, F)
    
```
